```python
import numpy as np
import pandas as pd
```

```python
original_file = "exams_info.csv"
three_classes_file = "traces100pc_exams_info.csv"
index_col = "id_exam"
```

```python
# Read exams info files with only classes 1, 2 and 3: traces 100%
three_classes_df = pd.read_csv(three_classes_file, index_col= index_col)
three_classes_index = three_classes_df.index
three_classes_df.head(10)
```

```python
# Read exams info file, original file with time information
exams_df = pd.read_csv(original_file, index_col = index_col)
exams_df = exams_df.reindex(three_classes_index, fill_value=False, copy=True) # Keep ids that are in three_classes_file
print(exams_df.isnull().sum() ==0) # It should print True, all ids in three_classes_file should be in original_file
exams_df.head(10)
```

```python
# Stats for time_to_first_appearance (weeks) -> class 3
exams_df.loc[exams_df["exam_class"] == 3, ["time_to_first_appearance"]].describe().transpose()
```

```python
# Stats for time_to_last_exam (weeks) -> class 1
exams_df.loc[exams_df["exam_class"] == 1, ["time_to_last_exam"]].describe().transpose()
```

```python
exams_df["time"] = exams_df["time_to_first_appearance"] + exams_df["time_to_last_exam"]
exams_df
```

```python
three_classes_df["time"] = exams_df["time"]
three_classes_df.tail(10)
```
